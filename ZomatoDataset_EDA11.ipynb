## Top 10 Most Common Cuisines

```python
plt.figure(figsize= (10,4))
df["Cuisines"].value_counts().head(10).plot(kind="bar",color="orange")
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.show()
```

## Observation :- Top 10 most popular Cuisines 
  The barchart shows that North Indian cuisine dominates the dataset, followed by North Indian- chinese and Chinese
   ,while street Food and Bakery-based cuisines have lower representation.
